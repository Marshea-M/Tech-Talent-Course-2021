```python
import altair as alt
import pandas as pd
```

```python
from vega_datasets import data
```

```python
autocars=pd.read_csv('imports85')
```

```python
brush = alt.selection_interval()

graph1=alt.Chart(autocars).mark_bar().encode(
    x='body-style:N',
    y='curb-weight:Q',
    color=alt.condition(brush,'Origin:N',alt.value('lightgray'))
).add_selection(
    brush    
)

graph1 
```

```python
graph2=alt.Chart(autocars).mark_line().encode(x='engine-size:Q',y='peak-rpm:Q',color=alt.value('red'))
graph2
```

```python
graph3=alt.Chart(autocars).mark_bar().encode(x='make:N',y='price:Q',color=alt.value('green'))
graph3
```

```python
brush = alt.selection_interval()

points2 = alt.Chart(autocars).mark_point().encode(
    x='curb-weight:Q',
    y='city-mpg:Q',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray')),
    tooltip=['make:N', 'fuel-type:N', 'horsepower:Q']
).add_selection(
    brush
)

bars2 = alt.Chart(autocars).mark_line().encode(
    y='curb-weight:Q',
    color='Origin:N',
    x='count(curb-weight):Q',
).transform_filter(
    brush
)

```

```python
#All of the graphs together for the visualisation
#I decided that a bar chart would be best for the first graph that displays the car's body style and weight because body style is a categorical value
#For the second graph I knew that it would be easier to see a correlation between engine size and peak revs per mile, if the chart was a line chart, and sometimes the peaks and falls of trends are not as clear with scatter diagrams
#For the third graph I knew that model and price would have great differences and that a bar chart would display distinct differences
#For the final two graphs I wanted to compliment or understand how one attribute is impacted by the weight of a car, so I displayed these graphs side-by-side to show the impact of weight and the frequency of just how many cars at that time were manufactured at certain weights

print("Data Visualisation of Automobiles from 1985 ") 
graph3
```
